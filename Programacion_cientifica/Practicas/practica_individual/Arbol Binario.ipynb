## Clases de pila y cola
Cargamos las clases que utilizaremos después para el recorrido del árbol. 

```python
class Pila():

    def __init__(self,tipo):
       self.__pila=list()
       self.tipo=tipo
       
    def estaVacia(self):
        return not self.__pila
    
    def cima(self):
        try:
            return self.__pila[-1]
        except:
            return None
         
    def apilar(self,elemento):
        
        if type(elemento)==self.tipo:
                self.__pila.append(elemento)
                return
        else: 
                raise TypeError
        
    def desapilar(self):
        try:
            return self.__pila.pop()
        except:
            None

class Cola:
    
    def __init__(self,tipo):
       self.__cola=list()
       self.tipo=tipo
       
    def estaVacia(self):
        return not self.__cola
    
    def primero(self):
        try:
            return self.__cola[0]
        except:
            return None
            
    def encolar(self,elemento):
        if type(elemento)==self.tipo:
                self.__cola.append(elemento)
                return
        else: 
                raise TypeError
        
    def desencolar(self):
        try:
            return self.__cola.pop(0)
        except:
            None
```

## Clase arbol binario 
En esta clase se han implementado también:

### Inorden Recursivo
##### Idea general. 
La idea para realizar la recursividad es la siguiente:
 1. Inicializamos una pila vacía P (llamada pila en el código) y una lista L.
 2. Inicializamos el nodo actual como la raiz primera.
 3. Hacemos un push del nodo actual, y redefinimos "nodo actual" como el hijo izquierdo, hasta que el hijo izquierdo sea None.
 4. Si el nodo actual es None y la pila no está vacía:
     <ol type="a">
         <li>Hacemos pop del últmo en entrar (LIFO)  y lo metemos en nuestra lista final L</li>
         <li>Forzamos a que el nodo actual sea el hijo derecho</li>
         <li>Repetimos el paso 3</li>
     </ol>
 5. Cuando el nodo actual sea None y la pila esté vacía, habremos terminado.
 
 
### Preorden Recursivo
##### Idea general. 
La idea para realizar la recursividad es la misma que para inorden, pero en este caso hemos cambiado el paso donde se añaden elementos a la pila.

### Recorrido amplitud
##### Idea general. 
 1. Inicializamos una cola vacía C (llamada cola en el código) y una lista L.
 2. Inicializamos el nodo actual como la raiz primera y lo encolamos.
 3. Mientras la cola no esté vacia:
 <ol type="a">
         <li>Desencolamos y establecemos el nodo_actual. </li>
         <li>Si está vacío, hacemos break. En caso de no estarlo, lo añadimos a la lista.</li>
         <li>Encolamos el hijo izquierdo y el hijo derecho.</li>
     </ol>
 4. Repetimos hasta que la cola esté vacía.
 
 ### Complejidad
 En los tres casos, el orden de complejidad es de o(n) porque ha de recorrer las n hojas que tenga el árbol una sola vez. 

```python
class ArbolBinarioOrdenado:
    
    def __init__(self):
        """constructor, los atributos no son privados"""
        self._raiz=None
        self._arbolIzdo=None
        self._arbolDcho=None
        
        
    def raiz(self):
        """Operacción de consulta: funcion que devuelve la raiz"""
        return self._raiz
    
   
    def arbolIzdo(self):
        """Operación de consulta: devuelve el subarbol izquierdo"""
        return self._arbolIzdo
    
    def arbolDcho(self):
        """Operación de consulta: devuelve el subarbol derecho"""
        return self._arbolDcho   
    
    def estaVacio(self):
        """Operación de consulta: devuelve el subarbol derecho"""
        return self._raiz==None
    
    def insertarElem(self,elemento):
        """Operación de inserción, se inserta el elemento  de forma ordenada"""
        if self.estaVacio():
            self._raiz=elemento
            self._arbolIzdo=ArbolBinarioOrdenado()
            self._arbolDcho=ArbolBinarioOrdenado()
        elif elemento<=self._raiz:
            self._arbolIzdo.insertarElem(elemento)
        elif elemento>self._raiz:
            self._arbolDcho.insertarElem(elemento)
        else:
            None
            
    def tieneElemento(self,elemento):
        """Operación de consulta: Comprueba si existe un  elemento en el árbol"""
        if self.estaVacio():
            return False
        elif self._raiz==elemento:
            return True
        elif elemento<self._raiz:
            return self._arbolIzdo.tieneElemento(elemento)
        else:
            return self._arbolDcho.tieneElemento(elemento)
        
    def numElementos(self):
        """Operación de consulta: Devuelve el númerro de elementos que hay en el árbol"""
        if self.estaVacio():
            return 0
        else:
            return 1+self._arbolIzdo.numElementos()+self._arbolDcho.numElementos()
               
    def eliminarArbol(self):
        """Operación de eliminación"""
        if not self.estaVacio():
            self._arbolIzdo.eliminarArbol()
            self._arbolDcho.eliminarArbol()
            self._raiz=None    
    
    def numHijos(self):
        if not self._arbolIzdo.estaVacio() and not self._arbolDcho.estaVacio():
            return 2
        else:
            return 1
        
    
    def noTieneHijos(self):
        """Comprueba si no  tiene hijos"""
        return self._arbolIzdo.estaVacio() and self._arbolDcho.estaVacio()

    
    def hijoUnico(self):
        """Devuelve el hijo único de un nodo que sólo tiene un hijo"""
        return self._arbolIzdo if self._arbolDcho.estaVacio() else self._arbolDcho
    
    
    def obtenerMasIzda(self,padre):
        """Obtiene el hijo más a la izquierda de un nodo"""
        
        if not self._arbolIzdo.estaVacio():
            
            return self._arbolIzdo.obtenerMasIzda(self._arbolIzdo)

        return [self,padre]
    
    
    def eliminarRaizConDosHijos(self):
        """Elimina un nodo con  dos hijos"""
       
        aux=self._arbolDcho.obtenerMasIzda(self)
       
        
        if not aux[0].estaVacio():
            
            self._raiz=aux[0]._raiz
            
            aux[0].eliminarElem(aux[0]._raiz,aux[1])
            
    
    def eliminarElem(self,elemento,padre=None):
        """Funcion que elimina  un elemento de un árbol"""
        if not self.estaVacio():
            
            if self._raiz==elemento:
              
                if self.noTieneHijos():
                   
                    if padre==None:
                        self._raiz=None
                    elif  padre._arbolIzdo._raiz==elemento:
                        padre._arbolIzdo._raiz=None
                    else:
                        padre._arbolDcho._raiz=None
                elif self.numHijos()==1:
                     
                     
                    if padre._arbolIzdo._raiz==elemento:
                        
                         padre._arbolIzdo=self.hijoUnico()
                    else:
                       
                        padre._arbolDcho=self.hijoUnico()
                    
                else:
                    
                    self.eliminarRaizConDosHijos()
            elif self._raiz>elemento:
                self._arbolIzdo.eliminarElem(elemento,self)
            else:
                self._arbolDcho.eliminarElem(elemento,self)
                
    def inOrden(self):
        """Recorrido en inorden, hijo izdo, raiz, hijo derecho"""
        l=[]
        if not self._arbolIzdo.estaVacio():
            l+=self._arbolIzdo.inOrden()
        l.append(self._raiz)
        if not self._arbolDcho.estaVacio():
            l+=self._arbolDcho.inOrden()

        return l
    
    def preOrden(self):
        """Recorrido en preorden, raiz, hijo izquierdo, hijo derecho"""
        l=[]
        l.append(self._raiz)
        if not self._arbolIzdo.estaVacio():
            l+=self._arbolIzdo.preOrden()
        if not self._arbolDcho.estaVacio():
            l+=self._arbolDcho.preOrden()

        return l
        
    def inorden_recursivo(self,tipo):
        # Recorrido en inorden, hijo izquierdo - raiz - hijo derecho 
        # Forzamos a que nodo_actual sea self para poder iterar por él.
        nodo_actual = self
        pila = Pila(tipo)
        lista = []
        while True:
            # Primero iremos a la hoja más a la izquierda
            if nodo_actual._raiz is not None:
                # Añadimos el nodo actual a la pila en la forma "raiz_del_arbol":"subarbol"
                # después forzamos que nodo_actual pase a la izquierda
                pila.apilar(nodo_actual)
                nodo_actual = nodo_actual._arbolIzdo
                
            # En este elif solo entrará si "nodo_actual" es None, es decir cuando el arbol_izdo anterior sea None,
            # pues python ejecutará en orden de cascada.
            
            elif(not (pila.estaVacia())):
                # Si la pila no está vacía, forzamos que el nodo_actual sea el último de la pila (LIFO)
                # para después añadirlo a nuestra lista. 
                nodo_actual = pila.desapilar()
                lista.append(nodo_actual._raiz)
                
                # Como ya hemos recorrido la raiz y su árbol izquierdo, toca su árbol derecho.
                nodo_actual = nodo_actual._arbolDcho
            else:
                return lista
                break
                
    def preorden_recursivo(self,tipo):
        nodo_actual = self
        pila = Pila(tipo)
        lista = []
        while True:
            if nodo_actual._raiz is not None:
                lista.append(nodo_actual._raiz)
                pila.apilar(nodo_actual)
                nodo_actual = nodo_actual._arbolIzdo
            elif(not (pila.estaVacia())):
                nodo_actual = pila.desapilar()
                nodo_actual = nodo_actual._arbolDcho
            else:
                return lista
                break

    def recorrido_amplitud(self,tipo):
        # Creamos la cola y listas vacías
        cola=Cola(tipo)
        lista=[]
        # Encolamos la raiz
        nodo_actual = self
        cola.encolar(nodo_actual)

        while not cola.estaVacia():
            # Añadimos la raiz a la lista y la quitamos de la cola, usando un auxiliar para poder recorrer los hijos
            nodo_actual=cola.desencolar()
            lista.append(nodo_actual._raiz)
            #Comprobamos que la raiz no sea none
            if (nodo_actual._raiz is None):
                break
            # Encolamos el hijo izquierdo
            if nodo_actual._arbolIzdo._raiz is not None:
                cola.encolar(nodo_actual._arbolIzdo)
                # Encolamos el hijo derecho
            if nodo_actual._arbolDcho._raiz is not None:
                cola.encolar(nodo_actual._arbolDcho)

        return lista
```

## Codigo del algoritmo de Huffman 
El algoritmo de Huffman se usa para la construcción de un código Huffman asociado a una cadena de caracteres. Para la implementación se debe contar con un diccionario que asocia a cada símbolo su frecuencia.

Para ello usaremos varias funciones:
#### get_freq
Me devuelve la frecuencia en una lista de palabras o frase.
#### make_tree
Usamos esta función para crear el árbol de Huffman con el que podremos hacer el encode de los textos. Notemos que no es un árbol ordenado como los anteriores.
#### make_code
Función recursiva (basda en las vistas en clase) para la obtención del código para cada letra.
#### encode_huffman
Función que utiliza las funciones anteriores para construir una cadena encodeada, de la cadena original que se le de.
